```python
for split_type in ['train','test','val']:
  if os.path.isdir(f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means/DETR/CT_DETR_kmeans/{split_type}') == False:
    os.makedirs(f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means/DETR/CT_DETR_kmeans/{split_type}')


  for cancer_type in ['lung_cancer','lymphoma','melanoma']:
  #for cancer_type in ['negative']:
    source_path = f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means/CT_{cancer_type}_k_means'
    #source_path = f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/negative/CT/data'
    destination_path = f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means/DETR/CT_DETR_kmeans/{split_type}'

    sampled_files = master_list[master_list['cancer_type']==cancer_type]
    sampled_files = sampled_files[sampled_files['split']==split_type]['file_name'].values
    for f in sampled_files:
      shutil.copyfile(source_path+f'/{f}', destination_path+f'/{f}')

```

```python
print("Done")
```

```python
labels = pd.read_csv('/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means2/SUV_labels.csv') # this is where all the labels are stored
for split_type in ['train','test','val']:
  files = master_list[master_list['split']==split_type]['file_name'].values
  split_labels = labels[labels['img_filename'].isin(files)]
  split_labels.to_csv(f'/content/drive/MyDrive/Capstone_GE_DSI_CV_Project/preprocessed_data/k_means/DETR/SUV_DETR_kmeans/{split_type}.csv')
```
