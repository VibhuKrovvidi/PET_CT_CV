```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
```

```python
image_name = 'stinkbug.png'

def draw_bb_yolo(bb_ground_truth, bb_predicted, image_name):
  ''' Input format: The inputs should be in pixel values, unscaled
  bb_ground_truth: list(x_center,y_center,w,h)
  bb_predicted: list(x_center,y_center,w,h)
  image_name: string
  '''
  im = Image.open(image_name)
  fig, ax = plt.subplots()
  ax.imshow(im)
  #ground truth - red
  rect = patches.Rectangle( (bb_ground_truth[0]-(bb_ground_truth[2]/2), bb_ground_truth[1]-(bb_ground_truth[3]/2) ), bb_ground_truth[2], bb_ground_truth[3], linewidth=1, edgecolor='r', facecolor='none')
  ax.add_patch(rect)
  #predicted - blue
  rect = patches.Rectangle( (bb_predicted[0]-(bb_predicted[2]/2), bb_predicted[1]-(bb_predicted[3]/2) ), bb_predicted[2], bb_predicted[3], linewidth=1, edgecolor='b', facecolor='none')
  ax.add_patch(rect)

  plt.show()

draw_bb_yolo([100,100,50,50], [200,200,20,20], image_name)
```
